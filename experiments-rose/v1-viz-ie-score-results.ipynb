```python
from matplotlib import pyplot as plt
import seaborn as sns

def visualise_ie_scores(prompt, ie_scores, results_desc = ['prompt_type', 'jailbreak_result']): 
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.scatterplot(x=range(len(ie_scores)), y=ie_scores, color='b', s=20, ax=ax)

    # Add title and subtitle
    plt.title(f'ie scores for {results_desc[0]}', fontsize=14)
    plt.suptitle(f'{results_desc[1]}', fontsize=8, fontweight='bold')

    # Add axis labels and descriptions
    plt.xlabel('prompt token index', fontsize=12)
    plt.ylabel('ie score', fontsize=12)
    ax.set_xlim(-0.5, len(ie_scores) - 0.5)  # Set x-axis limits for better spacing
```

```python
# Open the JSON file
with open('data/harmful_data_w_ie_score.json', 'r') as json_file:
    # Load the JSON data into a Python dictionary
    harmful_data = json.load(json_file)

# Open the JSON file
with open('data/adversarial_data_w_ie_score.json', 'r') as json_file:
    # Load the JSON data into a Python dictionary
    adv_data = json.load(json_file)

# Access the data
print(adv_data)
```
